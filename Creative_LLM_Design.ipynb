## Project Overview & Inspiration
This creative project was inspired by my internship experience at an AIGC department specializing in cultural heritage. During the summer, I worked with staff at the Hunan Provincial Museum in China, where I was given access to several artifact storage rooms. Inside were tens of thousands of items that had never been publicly displayed, from exquisite bronzeware and ceramics to quirky everyday objects from the Ming and Qing dynasties. But, because of space limitations and curatorial decisions, many of these artifacts remain sealed away from public view.

That experience made a deep impression on me. I kept thinking: why shouldn’t these lesser-known artifacts also get a chance to “meet” the audience?

Later in my internship, I worked with AI tools to extract information from museum publications and generate question-answer pairs about the artifacts. This process made me realize that AI wasn’t just useful for organizing heritage data — it could potentially help tell these objects’ stories. Through the course Artificial Intelligence for the Arts & Humanities, I developed a deeper understanding of how large language models (LLMs) work, which led to my project idea: What if we could give silent artifacts a voice?

This became the core concept of my design — a human-AI interactive platform powered by a large language model that lets museum visitors explore a virtual artifact storage room, where even non-exhibited items can “speak” and share their stories.

## Interaction Design Workflow
The system architecture includes three main components: a large language model (LLM), a cultural heritage artifact database, and a user interface.

### Step 1: User Entry
- Users enter the Virtual Artifact Storage Room through a web platform or an interactive display within the museum. Museums could set up small immersive spaces with projection screens to enhance the experience.

- Visitors can explore a visualized 3D map of the storage rooms, divided into categories such as bronzeware, ceramics, folk items, etc. They can click on any artifact they are curious about.

### Step 2: Activate an Artifact
- Each artifact comes with basic metadata (image, description, era, material). This information is used as input to the LLM.

- Once selected, the system opens a “digital artifact card.” The LLM generates a first-person introduction like:
"Hello, I’m a medicine jar from the Ming dynasty. I used to sit quietly in an herbal shop in a southern town."

### Step 3: Deep Interaction
- Behind each artifact is a mini LLM corpus that allows the user to ask follow-up questions like: 

  "Why are you shaped like this?"

   "Who did you belong to?"

   "Do you have any stories to share?"
- The LLM processes the input and generates personalized, conversational responses using relevant semantic knowledge.

- Contextual conversation is supported, allowing natural back-and-forth dialogue.

### Step 4: Extended Features
- Users can save their favorite artifacts to a personal memory book and export conversations as PDFs, postcards, or even creative illustrated booklets.

- The system recommends similar artifacts to build connections across different pieces.

- Users can provide feedback on their experience, helping improve recommendations.

- In the future, the system could integrate with AI image generators to allow users to create personalized artistic interpretations of artifacts.

Alternatively, users can type natural language prompts like:
“Are there any folk ceramics from the Song Dynasty?”
“Do you have any artifacts related to women’s daily life?”
“I’m interested in mythical stories—show me something!”
The LLM would understand the query, retrieve relevant items, and present them in a narrative, first-person style.

## Value of the Design
Traditional museum plaques only offer fixed, simplified information. With a large language model, users can freely ask layered, follow-up questions and receive rich, personalized responses. This lowers the barrier for all kinds of audiences, from kids and teens to casual visitors and culture lovers — even those without historical or academic background.

Rather than passively viewing exhibitions, users can now “talk” with artifacts, build their own narratives, and explore stories they genuinely care about.

## Reflection: Cultural and Social Impact
**Positive Impacts:**

One of the most meaningful aspects of this project is its potential to decentralize cultural knowledge. Many artifacts that never make it to the gallery floor can now be seen and appreciated. With digital access, people in classrooms, libraries, or rural areas can interact with heritage objects anytime, anywhere.


**Potential Challenges:**

Of course, there are risks. First, LLMs may generating inaccurate or misleading information. If unchecked, this could damage the integrity of cultural interpretation. To ensure accuracy, the system should draw on a museum-approved artifact database as its foundational source, supplemented by expert-reviewed descriptive texts for fine-tuning.

Second, there's a concern that if users can experience everything online, will they still visit the museum? My view is that this system should be designed as a complement, not a replacement. Features like immersive screens, VR, and interactive zones in physical museums can make the in-person experience even more magical — with online exploration becoming the spark that draws people in.
